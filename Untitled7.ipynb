```python
# ===== STEP 1: Install =====
!pip install openai-whisper -q
!apt install ffmpeg -y -q

# ===== STEP 2: Upload audio files =====
from google.colab import files
print("👉 Click 'Choose Files' and select 3-4 audio files (mp3/wav)")
uploaded = files.upload()

print("\n✅ Uploaded files:", list(uploaded.keys()))

# ===== STEP 3: Load Whisper model =====
import whisper
model = whisper.load_model("base")
print("✅ Model loaded successfully!")

# ===== STEP 4: Filter valid audio files =====
audio_files = [f for f in uploaded.keys() if f.lower().endswith(('.mp3', '.wav', '.m4a'))]
print("✅ Audio files to process:", audio_files)

if len(audio_files) == 0:
    print("⚠️ WARNING: No valid audio files found! Check file names/extensions above.")

# ===== STEP 5: Transcribe function =====
def transcribe_audio(file_path):
    result = model.transcribe(file_path)
    print(f"\n📁 File: {file_path}")
    print(f"🌐 Detected Language: {result['language']}")
    print(f"📝 Full Transcription: {result['text']}\n")

    for seg in result["segments"]:
        confidence = round((1 + seg["avg_logprob"]) * 100, 2)
        confidence = max(0, min(confidence, 100))
        print(f"  [{seg['start']:.1f}s - {seg['end']:.1f}s] \"{seg['text'].strip()}\" (confidence: {confidence}%)")

    return result

# ===== STEP 6: Run transcription on all files =====
results = {}
for audio_file in audio_files:
    results[audio_file] = transcribe_audio(audio_file)

print("\n✅ Total results generated:", len(results))

# ===== STEP 7: Save to file =====
with open("transcriptions_output.txt", "w") as f:
    for file, result in results.items():
        f.write(f"File: {file}\n")
        f.write(f"Language: {result['language']}\n")
        f.write(f"Transcription: {result['text']}\n")
        f.write("-" * 50 + "\n")

print("✅ Saved to transcriptions_output.txt")

# ===== STEP 8: Download =====
from google.colab import files as colab_files
colab_files.download("transcriptions_output.txt")
```
